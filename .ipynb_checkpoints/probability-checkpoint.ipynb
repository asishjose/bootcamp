```python
def conditional_probability(P_A_and_B, P_B):
    return P_A_and_B / P_B

# Number of face cards = 12, Number of Kings = 4
P_A_and_B = 4 / 52  # Probability of drawing a King and it's a face card
P_B = 12 / 52  # Probability of drawing a face card
P_A_given_B = conditional_probability(P_A_and_B, P_B)
print(f"Conditional Probability: {P_A_given_B}")
```

BAYES THEOREM

```python
def bayes_theorem(P_D_given_A, P_A, P_D):
    return (P_D_given_A * P_A) / P_D

P_A = 0.7  # Probability of choosing Machine A
P_B = 0.3  # Probability of choosing Machine B
P_D_given_A = 0.01  # Probability of defect from Machine A
P_D_given_B = 0.02  # Probability of defect from Machine B

P_D = (P_D_given_A * P_A) + (P_D_given_B * P_B)  # Total probability of defect
P_A_given_D = bayes_theorem(P_D_given_A, P_A, P_D)
print(f"Probability Machine A produced the defective item: {P_A_given_D}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate rolling a fair dice 1000 times
rolls = np.random.choice([1, 2, 3, 4, 5, 6], size=1000)

# Plot the distribution
plt.hist(rolls, bins=6, edgecolor='black', density=True)
plt.title('Probability Distribution of Rolling a Fair Dice')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

```python
# Generate a random sample from a normal distribution
normal_sample = np.random.normal(loc=0, scale=1, size=1000)

# Plot the histogram
plt.hist(normal_sample, bins=30, edgecolor='black', density=True)
plt.title('Normal Distribution (Mean=0, Std=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Simulate 1000 trials of 10 coin flips (success probability = 0.5)
binomial_trials = np.random.binomial(n=10, p=0.5, size=1000)

# Plot the distribution
plt.hist(binomial_trials, bins=11, edgecolor='black', density=True)
plt.title('Binomial Distribution: 10 Coin Flips')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()
```
