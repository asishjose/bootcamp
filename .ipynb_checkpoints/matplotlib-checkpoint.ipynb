# MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
```

```python
y=np.arange(11,21)
```

```python
#scatter plot
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
plt.show()
```
