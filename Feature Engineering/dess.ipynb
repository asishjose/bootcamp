```python
x.fillna({'age':0},inplace=True)
q1=np.percentile(x['age'],25)
q3=np.percentile(x['age'],75)
iqr=q3-q1
iqr
```

#### 

```python
x['age'].median()
```

```python
lower_bound=(q1-1.5)*iqr
upper_bound=(q1+1.5)*iqr
```

```python
lower_bound, upper_bound
```

```python
v = x[(x['age']<lower_bound)|(x['age']>upper_bound)]
```
