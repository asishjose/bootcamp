```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = {
    'Height': [150, 160, 170, 180, 190],
    'Weight': [45, 60, 70, 80, 90]
}
df = pd.DataFrame(data)

# Pearson correlation
pearson_corr = df['Height'].corr(df['Weight'])
print(f"Pearson Correlation: {pearson_corr}")

# Visualization of the relationship
sns.scatterplot(x='Height', y='Weight', data=df)
plt.title('Height vs Weight')
plt.show()
```

```python
# Example dataset
data = {
    'Subject1': [90, 85, 80, 95, 88],
    'Subject2': [85, 90, 80, 92, 86]
}
df = pd.DataFrame(data)

# Spearman correlation
spearman_corr = df['Subject1'].corr(df['Subject2'], method='spearman')
print(f"Spearman Correlation: {spearman_corr}")

# Visualization (Optional)
sns.scatterplot(x='Subject1', y='Subject2', data=df)
plt.title('Subject1 vs Subject2')
plt.show()
```

```python
# Example dataset with more variables
data = {
    'Height': [150, 160, 170, 180, 190],
    'Weight': [45, 60, 70, 80, 90],
    'Age': [20, 22, 24, 26, 28],
    'Score': [88, 92, 85, 89, 91]
}
df = pd.DataFrame(data)

# Correlation matrix
corr_matrix = df.corr()
print(corr_matrix)

# Visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
