## Seaborn
### is a Python data visualization library built on top of matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

## Box Plot
Used to visualize the distribution, median, and outliers of numeric data.

```python
sns.boxplot(x='day',y='total_bill',data=df)
plt.title("Box Plot: Total Bill by Day")
plt.show()
```

## Violin Plot
Combines a box plot with a KDE (Kernel Density Estimate) to show distribution and probability density.

```python
sns.violinplot(x='day', y='total_bill',data=df, inner='box')
plt.title("Violin Plot: Total Bill by Day")
```

## Pair Plot
Used to visualize pairwise relationships in a dataset â€” great for EDA!

```python
sns.pairplot(df, hue='sex')
plt.suptitle('Pairplot: Tips Dataset', y=1.02)
```
