## Univariate  Analysis

```python
plt.hist(x='mpg',data=mtcars_data)
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('MPG Distribution')
```

```python
sns.histplot(mtcars_data,x = 'mpg',bins = 10, kde=True,edgecolor=None)
plt.xlabel('MPG')
```

```python
import seaborn as sns
sns.countplot(data=mtcars_data,x='cyl', hue = 'cyl')
plt.title('Cylinder Distribution')
plt.show()
```

```python
plt.boxplot(x='mpg',data=mtcars_data)
plt.show()
```

```python
mtcars_data['cyl'].value_counts()
```

```python
plt.pie(x=mtcars_data['cyl'].value_counts(), labels=['8cyl','4cyl','6cyl'])
plt.show()
```

## Bivariate Analysis

#### Barplot - 1 Discrete Vs 1 Continous

```python
sns.barplot(x='gear', y='mpg', data=mtcars_data, hue='gear')
```

#### ScatterPlot - Cont Vs Cont

```python
sns.scatterplot(x='hp', y='mpg', data=mtcars_data)
plt.title('HP vs MPG')
```

#### Boxplot - Distribution + Outliers

```python
sns.boxplot(x='cyl', y='mpg', data=mtcars_data, hue='cyl')
```
