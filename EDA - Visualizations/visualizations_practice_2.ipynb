```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('tips')
data.head()
```

### scatter plot

```python
plt.scatter(data['day'],data['tip'])
```

```python
plt.scatter(data['day'],data['tip'], c=data['size'], s=data['total_bill'])

plt.colorbar()
plt.title = ('Scatter Plot')
plt.xlabel('Day')
plt.ylabel('Tip')
```

color (c) is determined by size

size (s) of the bubbles is proportional to total_bill

plt.colorbar() adds a color legend for size

### Line Chart

```python
plt.plot(data['tip'])
plt.plot(data['size'])

#plt.title('Scatter_Plot')
plt.xlabel('Day')
plt.ylabel('Tip')
```

### Multiple Charts

```python
import plotly.graph_objects as px
import pandas as pd

data = sns.load_dataset('tips')
plot = px.Figure(data=[px.Scatter(y=data['tip'],mode='lines',)])
 
plot.update_layout(xaxis=dict(rangeselector=dict(buttons=list([dict(count=1,step="day",
                    stepmode="backward"),])),rangeslider=dict(visible=True),)) 
plot.show()
```
