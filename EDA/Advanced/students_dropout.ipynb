```python
import numpy as np

# Add missing values to numeric column
df.loc[df.sample(frac=0.1).index, 'age'] = np.nan  # 10% missing in 'age'

# Add missing values to categorical column
df.loc[df.sample(frac=0.1).index, 'scholarship'] = np.nan  # 10% missing in 'scholarship'
```
