# apply()

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns').head()
```

```python
def stars(row):
    if (row.country == 'Canada') | (row.points >= 95):
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1
star_ratings = reviews.apply(stars, axis=1)
```
