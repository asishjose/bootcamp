```python
from sklearn.feature_selection import VarianceThreshold
import numpy as np

# Example dataset (features with low and high variance)
data = np.array([[1, 0, 0],
                 [1, 1, 0],
                 [1, 0, 1],
                 [1, 1, 1]])

# Apply VarianceThreshold with a threshold of 0.2
selector = VarianceThreshold(threshold=0.2)
filtered_data = selector.fit_transform(data)

print(filtered_data)
```
