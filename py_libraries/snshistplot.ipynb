```python
import pandas as pd

df = pd.DataFrame({
    'Scores': [85, 70, 90, 88, 76, 94, 73, 95],
    'Category': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B']
})

sns.histplot(df, x='Scores', hue='Category', multiple='stack', kde=True)
plt.show()
```

```python

data = [7, 8, 9, 10, 11, 12, 12, 13, 14, 19]
df = pd.DataFrame(data, columns=['Numbers'])
sns.boxplot(data=data)
```

```python

df = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'Values': [10, 12, 10, 50, 9, 8, 9, 11, 100]  # 50 and 100 are outliers
})

sns.boxplot(x='Category', y='Values', data=df, hue='Category', palette='Set2', legend=False)
plt.title("Boxplot with 'hue' for Categories")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram, KDE, and normal distribution PDF
sns.histplot(data, kde=True, stat='density', color='skyblue')
plt.title("KDE and Normal Distribution PDF")
plt.show()
```
